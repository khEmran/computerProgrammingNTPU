```python
#defining class with construcor and methods and executioning them
class Vehicle:
    def __init__(self, name="",kind="car", color="",value=100.00):
        self.name = name
        self.kind = kind
        self.color = color
        self.value = value
    def description(self):
        return f"{self.name} is a {self.color} {self.kind} worth ${self.value:,.2f}."

car1 = Vehicle(name="HONDA", kind="Convertible", color="Silver", value=8500.00)
car2 = Vehicle(name="AUDI", kind="Van", color="Red", value=79000.00)

print(car1.description())
print(car1.name)
print(car2.description())
print(car2.name)
```

```python
#exploring try-except-else-finally keywords, error handling
try:
    price = float(input("Enter the stock price:"))
    shares = int(input("Enter the number of shares (e.g. 2):"))
    total = price * shares
except Exception as e:
    print("Exception error:", str(e))
else:
    print("Share Total Value is:", total)
finally:
    print("Thank You.")
```

Option A – Healthcare 
Create a Patient Record class 

• patient name  
• age  
• diagnosis  
• save to file  
• handle invalid age input 

```python
class PatientRecord: 
    def __init__(self, name="", age="", diagnosis=""): 
        self.name = name 
        self.age = age 
        self.diagnosis = diagnosis 
    def save_record(self): 
        with open("patient_record.txt", "w") as f: 
            f.write(f"{self.name}, {self.age}, {self.diagnosis}") 

try: 
    p = PatientRecord(name="Mr. X", age= 40, diagnosis="Hypertension") 
    p.save_record() 
    print("Patient record saved.") 
except Exception as e: 
    print("Error:", e) 
```
