```python
#defining class and initializing it with constructor and also exploring method
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age= age
    def myFunc(self):
        print("Hello, I am "+self.name)
        
p1 = Person("Emran", 24)
p1.myFunc()
print(p1.name)
print(p1.age)
```

```python
#defining class with construcor and methods and executioning them
class Vehicle:
    def __init__(self, name="",kind="car", color="",value=100.00):
        self.name = name
        self.kind = kind
        self.color = color
        self.value = value
    def description(self):
        return f"{self.name} is a {self.color} {self.kind} worth ${self.value:,.2f}."

car1 = Vehicle(name="HONDA", kind="Convertible", color="Silver", value=8500.00)
car2 = Vehicle(name="AUDI", kind="Van", color="Red", value=79000.00)

print(car1.description())
print(car1.name)
print(car2.description())
print(car2.name)
```
